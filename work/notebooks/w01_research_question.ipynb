## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load starter dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Quick information
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
print("Clients:", df["client_id"].nunique())

# Supporting numbers
print("Average CTR (%):", round(df["ctr"].mean(), 2))
print("Average Position:", round(df["avg_position"].mean(), 2))
print("Average Content Age (days):", round(df["content_age_days"].mean(), 2))
```

The starter dataset contains 30,000 content pages from 32 clients with 44 features. The average CTR is 0.51%, the average position is 16.34, and the average content age is about 256.17 days. These numbers suggest that there is enough variation in the data to investigate which pages may benefit from a content refresh.
